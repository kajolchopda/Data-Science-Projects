In the describe function we see that many columns with Numeric values are present in Object data type like Interest Rate,Fico Range and few others.

- This is because the columns have numeric value along with some symbol like (%,-) and therefore we find them with object datatype.

- In the further process we will convert them into numeric values to perform Linear Regression.

```python
#Creating target variable in test column

test['Interest.Rate']=''

train.shape,test.shape
```

```python
#Creating new column, in both dataset so that when we combine we get to know data data belongs to train or test.

train['Data_name']='train'
test['Data_name']='test'
```

```python
#combining both data

combined=pd.concat([train,test],axis=0,ignore_index=True,sort=False)

combined.shape
```

```python
combined.info()
```

```python
#We will drop the Id,Amount Funded by Investor column because they are not useful for our prediction model.
#To create a model we consider only past/previous data therefore we drop Amount Funded by Investor column,its a future variable

combine=combined.drop(['ID','Amount.Funded.By.Investors'],axis=1)
```

```python
#Removing % sign from Interest Rate,Debt.To.Income.Ratio column

combine['Interest.Rate']=combine['Interest.Rate'].str.replace('%','')
combine['Debt.To.Income.Ratio']=combine['Debt.To.Income.Ratio'].str.replace('%','')

#converting it to numeric datatype
combine['Interest.Rate']=pd.to_numeric(combine['Interest.Rate'],errors='coerce')
combine['Debt.To.Income.Ratio']=pd.to_numeric(combine['Debt.To.Income.Ratio'],errors='coerce')
```

```python
#Also converting few other columns into numeric

for col in ['Open.CREDIT.Lines','Revolving.CREDIT.Balance','Amount.Requested']:
    combine[col]=pd.to_numeric(combine[col],errors='coerce')
```

```python
combine.info()
```

```python
#Splitting FICO range,and then taking sum of both splitted value and dividing by 2

splitting=combine['FICO.Range'].str.split('-',expand=True).astype(float)

combine['FICO']=(splitting[0]+splitting[1])* 0.5

del combine['FICO.Range']
```

```python
combine['Employment.Length'].unique()
```

```python
#Cleaning the Employment Length variable

combine['Employment.Length']=combine['Employment.Length'].str.replace('years','')
combine['Employment.Length']=combine['Employment.Length'].str.replace('year','')

combine['Employment.Length']=np.where(combine['Employment.Length'].str[:2]=='10',10,combine['Employment.Length'])
combine['Employment.Length']=np.where(combine['Employment.Length'].str[0]=='<',0,combine['Employment.Length'])
combine['Employment.Length']=pd.to_numeric(combine['Employment.Length'],errors='coerce')
```

```python
combine.info()
```

```python
#Now selecting all Categorical column seperately

cat_cols=combine.select_dtypes(['object']).columns
cat_cols
```

```python
#We will not consider Data_name column as it is temporarily added

cat_cols=cat_cols[:-1]
```

```python
#Creating dummy variables
#Also checking that too many attributes are not created and control the dummies based on frequency

for col in cat_cols:
    freq=combine[col].value_counts()
    k= freq.index[freq>20][:-1]
    
    for cat in k:
        name=col+'_'+cat
        combine[name]=(combine[col]==cat).astype(int)
    del combine[col]
    print(col)
```

```python
#checking missing values

combine.isnull().sum()[combine.isnull().sum()!=0]
```

```python
#imputing Null vallues with mean

for col in combine.columns:
    if(col not in['Interest.Rate','Data_Name']) & (combine[col].isnull().sum()>0):
        combine.loc[combine[col].isnull(),col]=combine.loc[combine['Data_name']=='train',col].mean()
```

```python
combine.isnull().sum()[combine.isnull().sum()!=0]
```

- We have imputed the null values except for the target variable that had missing values for test dataset.

```python
#creating a copy of combine data
combine_copy=combine.copy()

#Now we will seperate the train and test data:
train=combine[combine['Data_name']=='train']
del train['Data_name']

test=combine[combine['Data_name']=='test']
del test['Data_name']
```

```python
del combine
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import Ridge,Lasso
from sklearn.model_selection import GridSearchCV
```

```python
train1,train2=train_test_split(train,test_size=0.2,random_state=2)
```

```python
#We have only used train data to impute the missing values

x_train1=train1.drop('Interest.Rate',axis=1)
y_train1=train1['Interest.Rate']
```

```python
lr=LinearRegression()
lr.fit(x_train1,y_train1)
```

```python
#B0 i.e our constant value
lr.intercept_
```

```python
#coeff of evry variable i.e B1,B2...

list(zip(x_train1.columns,lr.coef_))
```

```python
#predicting on train2 data

x_train2=train2.drop('Interest.Rate',axis=1)

#giving x_train2 data for prediction as we have removed targe variable.
predicting=lr.predict(x_train2)
predicting
```

```python
#checking the mean absolute error value
mean_absolute_error(train2['Interest.Rate'],predicting)
```

- The above value is calculated by finding the difference between the actual value and the predicted value.

We know the tentative performance,now we will check on the entire training set.

```python
x_train=train.drop('Interest.Rate',axis=1)
y_train=train['Interest.Rate']
```

```python
lr.fit(x_train,y_train)
```

```python
#Target variable is present in test data,we will first drop that attribute and then predict values
test=test.drop('Interest.Rate',axis=1)
```

```python
#Prediction for test data
test_prediction=lr.predict(test)
test_prediction
```

- Above cell have the values that are predicted for the Interest Rate in test dataset.

We will perform Hyperparameter Tuning for better solution.

```python
#In this we take 100 values ranging between 1-100
lambdas=np.linspace(1,100,100)

#for parameters selection alpha value is selected from lambdas value
params={'alpha':lambdas}

#for creating a model we use Ridge and set fit itercept as true
model=Ridge(fit_intercept=True)
```

```python
#Using GridSearchCV we do Hyperparameter Tuning for better optimal value
grid_search=GridSearchCV(model,param_grid=params,scoring='neg_mean_absolute_error',cv=10)

grid_search.fit(x_train,y_train)
```

```python
#checking our best estimator
grid_search.best_estimator_
```

```python
grid_search.cv_results_
```

```python
#After using GridSearchCV we will again check the prediction value for test data
test_prediction=grid_search.predict(test)

test_prediction
```

```python
#Checking the coefficients using Ridge model
ridge_model=grid_search.best_estimator_

ridge_model.fit(x_train,y_train)

#We will check the coefficient for all the columns
list(zip(x_train.columns,ridge_model.coef_))
```

# Lasso Regression

```python
lambdas=np.linspace(1,10,100)

params={'alpha':lambdas}

model=Lasso(fit_intercept=True)
```

```python
grid_searchcv=GridSearchCV(model,param_grid=params,cv=10,scoring='neg_mean_absolute_error')

grid_searchcv.fit(x_train,y_train)
```

```python
lasso_model=grid_searchcv.best_estimator_

lasso_model.fit(x_train,y_train)
```

```python
list(zip(train1.columns,lasso_model.coef_))
```

- We have performed Hyperparameter tuning using GridSearchCV on Lasso and Ridge Regression.
